```python
from sklearn.ensemble import BaggingRegressor, AdaBoostRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
```

```python
t1=train1['Purchase']
t1.head(1)
```

```python
t2=train1.drop(['Purchase','Product_ID'],axis=1)
t2.head()
```

```python
x_train, x_test, y_train, y_test= train_test_split(t2,t1,test_size=0.2, random_state=123)
```

```python
tree=DecisionTreeRegressor()
rf=RandomForestRegressor()
bagg=BaggingRegressor(base_estimator=tree)
boost=AdaBoostRegressor(base_estimator=tree)
```

```python
param_grid_tree={'criterion':['gini','entropy'],'max_depth':[3,5,7,9]}
param_grid_rf={'n_estimators':[100,200,500],'max_depth':[3,5,7,9]}
param_grid_bagg={'n_estimators':[100,200,500]}
param_grid_boost={'n_estimators':[100,200,500],'learning_rate':[0.0001,0.0003,0.0005]}
```

```python
best_model_tree=GridSearchCV(estimator=tree,param_grid=param_grid_tree,scoring=None,cv=5)
best_model_rf=GridSearchCV(estimator=rf,param_grid=param_grid_rf,scoring=None,cv=5)
best_model_bagg=GridSearchCV(estimator=bagg,param_grid=param_grid_bagg,scoring=None,cv=5)
best_model_boost=GridSearchCV(estimator=boost,param_grid=param_grid_boost,scoring=None,cv=5)
```

```python
best_model_tree.fit(X=x_train,y=y_train)
```
